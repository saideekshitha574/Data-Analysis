# Description About Dataset

# The dataset contains daily sub-basin-wise rainfall data from the India Meteorological Department (IMD) for July 2023. It consists of 2,185 records and 7 columns
Basin: The name of the major basin in which the sub-basin is located.
Subbasin: The specific sub-basin for which rainfall data is recorded.
Date: The date of the recorded rainfall (YYYY-MM-DD format).
Year: The year of the recorded data (2023).
Month: The month of the recorded data (July = 7).
Avg_rainfall: The average daily rainfall (in mm). Some missing values exist.
Agency_name: The name of the agency providing the data (IMD GRID MODEL).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

# Loading dataset


```python
file_path = "Daily_Sub-basin-wise_Rainfall_data_from_India_Meteorological_Department_Agency_during_July_2023.csv"
df = pd.read_csv(file_path)
df
```

# DATASET INFOO

```python
df.info()
```

#  COUNTING THE NULL VALUES

```python
df.isnull().sum()
```

#  FILLING THE NULL VALUES WITH ZEROS

```python
df=df.fillna(0)
df
```

# MAKING THE NO NULL VALUES ARE THERE
 

```python
 df.isnull().sum()
```

```python
df
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df
```

```python
# Drop rows where 'Date' conversion failed
df = df.dropna(subset=['Date'])
df
```

```python
# Fill missing rainfall values with the mean rainfall of respective sub-basins
df['Avg_rainfall'] = df.groupby('Subbasin')['Avg_rainfall'].transform(lambda x: x.fillna(x.mean()))
df
```

```python
# If still NaN, fill with overall mean rainfall
df['Avg_rainfall'].fillna(df['Avg_rainfall'].mean(), inplace=True)
df
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
df
```

```python
# Reset index
df.reset_index(drop=True, inplace=True)
df
```

```python
# Save cleaned data
df.to_csv("Cleaned_Rainfall_Data.csv", index=False)
print("Data cleaning complete. Cleaned file saved as 'Cleaned_Rainfall_Data.csv'")
```

# Data Visualization

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Avg_rainfall'], bins=30, kde=True, color='blue')
plt.title('Distribution of Average Rainfall')
plt.xlabel('Avg Rainfall (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Avg_rainfall'], color='skyblue')
plt.title('Boxplot of Average Rainfall')
plt.xlabel('Avg Rainfall (mm)')
plt.show()
```

# Bar chart of average rainfall per sub-basin

```python
df_grouped = df.groupby('Subbasin')['Avg_rainfall'].mean().sort_values(ascending=False)
plt.figure(figsize=(16, 6))
df_grouped.plot(kind='bar', color='teal')
plt.title('Average Rainfall per Sub-basin')
plt.xlabel('Sub-basin')
plt.ylabel('Avg Rainfall (mm)')
plt.xticks(rotation=90)
plt.show()

```

# Line plot of daily average rainfall

```python
df_date_grouped = df.groupby('Date')['Avg_rainfall'].mean()
plt.figure(figsize=(12, 6))
sns.lineplot(x=df_date_grouped.index, y=df_date_grouped.values, color='red')
plt.title('Daily Average Rainfall Over Time')
plt.xlabel('Date')
plt.ylabel('Avg Rainfall (mm)')
plt.xticks(rotation=45)
plt.show()
```

# Scatter plot of rainfall vs. date


```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['Date'], y=df['Avg_rainfall'], color='green', alpha=0.5)
plt.title('Scatter Plot of Rainfall Over Time')
plt.xlabel('Date')
plt.ylabel('Avg Rainfall (mm)')
plt.xticks(rotation=45)
plt.show()
```

# Heatmap of average rainfall by sub-basin and date


```python
df_pivot = df.pivot_table(values='Avg_rainfall', index='Subbasin', columns='Date', aggfunc=np.mean)
plt.figure(figsize=(15, 8))
sns.heatmap(df_pivot, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Average Rainfall by Sub-basin and Date')
plt.xlabel('Date')
plt.ylabel('Sub-basin')
plt.show()
```

```python
# Generate day-wise individual bar and pie charts
unique_dates = df['Date'].unique()
for date in unique_dates:
    daily_data = df[df['Date'] == date]  
# Bar Chart
plt.figure(figsize=(16, 6))
sns.barplot(x='Subbasin', y='Avg_rainfall', data=daily_data, palette='viridis')
plt.xticks(rotation=90)
plt.title(f"Rainfall Distribution on {date.strftime('%Y-%m-%d')}")
plt.xlabel("Sub-basin")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
# Basic Statistics
print("Basic Rainfall Statistics:")
print(df['Avg_rainfall'].describe())
df
```

```python
# Top 5 Wettest and Driest Days
daily_rainfall = df.groupby('Date')['Avg_rainfall'].sum()
top_wettest_days = daily_rainfall.nlargest(5)
top_driest_days = daily_rainfall.nsmallest(5)
print("Top 5 Wettest Days:")
print(top_wettest_days)
print("Top 5 Driest Days:")
print(top_driest_days)
df
```

```python
# Daily Rainfall Trend (Line Chart)
plt.figure(figsize=(12, 6))
plt.plot(daily_rainfall.index, daily_rainfall.values, marker='o', linestyle='-', color='b')
plt.xlabel("Date")
plt.ylabel("Total Rainfall (mm)")
plt.title("Daily Rainfall Trend")
plt.xticks(rotation=45)
plt.grid()
plt.show()
df
```

```python
# Subbasin-wise Rainfall Distribution (Box Plot)
plt.figure(figsize=(20, 16))
sns.boxplot(x='Subbasin', y='Avg_rainfall', data=df, palette='coolwarm')
plt.xticks(rotation=90)
plt.xlabel("Subbasin")
plt.ylabel("Avg Rainfall (mm)")
plt.title(" Subbasin-wise Rainfall Distribution")
plt.show()
df
```

# 1.Introduction
The dataset analyzed in this report consists of daily sub-basin-wise rainfall data obtained from the India Meteorological Department for July 2023. The analysis involves data cleaning, exploration, visualization, and statistical evaluation to understand rainfall distribution patterns.

The dataset underwent the following preprocessing steps:
->Date Conversion: The Datecolumn was converted to datetime format, and invalid values were removed.
->Missing Value Handling:Missing rainfall values were replaced with the mean rainfall for respective sub-basins. If sub-basin-level averages were unavailable, the overall dataset mean was used.
->Duplicate Removal: Duplicate records were eliminated to ensure data integrity.
->Index Resetting: The dataset index was reset after cleaning to maintain consistency.
# 3.Statistical Summary
- The dataset contains sub-basin-wise daily rainfall values.
- Descriptive statistics, including mean, median, minimum, and maximum rainfall, were calculated.
- The date with the highest recorded rainfall was identified.
